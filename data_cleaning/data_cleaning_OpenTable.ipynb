```python
import pandas as pd
import numpy as np
import re
```

```python
opentable_df = pd.read_csv('/Users/marianaarrieta/Documents/Data_Projects/london_restaurants/raw_data/opentable_data.csv')
```

```python
opentable_df.head()
```

```python
#Remove duplicates, if any 
opentable_df.drop_duplicates(subset = ['names'], inplace = True)

#Remove booking column 
opentable_df.drop(columns=['bookings'], inplace = True)

#Remove rows where there are no reviews
opentable_df = opentable_df[opentable_df['reviews'].notna()]

#Remove rows where the restaurant name contains "Permanently closed"
opentable_df = opentable_df[opentable_df["names"].str.contains("Permanently Closed") == False]

#Standarize caps
opentable_df['names'] = opentable_df['names'].str.lower()
opentable_df['names'] = opentable_df['names'].str.title()
```

```python
#Remove rows with irrelevant cuisines
opentable_df = opentable_df[opentable_df.cuisine != 'Bar / Lounge / Bottle Service']
opentable_df = opentable_df[opentable_df.cuisine != 'Unspecified']
opentable_df = opentable_df[opentable_df.cuisine != 'Wine Bar']
opentable_df = opentable_df[opentable_df.cuisine != 'Lounge']
opentable_df = opentable_df[opentable_df.cuisine != 'Afternoon Tea']
opentable_df = opentable_df[opentable_df.cuisine != 'Breakfast']
opentable_df = opentable_df[opentable_df.cuisine != 'Pub']
opentable_df = opentable_df[opentable_df.cuisine != 'Dessert']
opentable_df = opentable_df[opentable_df.cuisine != 'Creperie']
opentable_df = opentable_df[opentable_df.cuisine != 'Cafe Dining']
```

```python
cuisine_dict = {"Fish" : 'Seafood', "European": "Modern European", 
"Tapas / Small Plates": 'Spanish', "Contemporary British": "British", "Californian": "American", "Steakhouse": "Steak", 
"Dim Sum": "Chinese", "Pizza Bar": "Italian", "Contemporary European": "Modern European", "Lebanese": "Middle Eastern", 
"Turkish":"Middle Eastern", "Syrian": "Middle Eastern", "Contemporary French": "French", "Japanese (Yakiniku)": "Japanese", 
"Vegan": "Vegetarian / Vegan", "English": "British", "Contemporary Asian": "Modern Asian", "Japanese Speciality": "Japanese", 
"Global, International": "International", "Pizzeria": "Italian", "Sicilian": "Italian", "Argentinian": "Steak", 
"Sushi": "Japanese", "Continental":"International"}
```

```python
opentable_df =opentable_df.replace({"cuisine": cuisine_dict})
```
